```python
import matplotlib.pyplot as plt
from matplotlib_inline import backend_inline

# 创建示例数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 8, 13]

# 创建折线图
plt.plot(x, y, label='线条1', color='blue', linestyle='-', marker='o')

# 添加标题和标签
plt.title('示例折线图')
plt.xlabel('X轴标签')
plt.ylabel('Y轴标签')

# 添加图例
plt.legend()
plt.grid(ls=':')
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt

# 创建示例数据
x = [1, 10, 100, 1000, 10000]
y = [1, 2, 3, 4, 5]

# 绘制折线图
plt.plot(x, y, label='数据')

# 设置 x 轴刻度为对数缩放
plt.xscale('log')

# 添加标题和标签
plt.title('示例折线图 with 对数刻度')
plt.xlabel('X轴')
plt.ylabel('Y轴')

# 显示图例
plt.legend()

# 显示图形
plt.show()
```
